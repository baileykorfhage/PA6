```python
import math

class calc():
    def __init__(self,input1,input2):
        self.input1=input1
        self.input2=input2
        
    def addition(num1,num2):
        print (f'\nSum: {num1+num2}')
    def subtract(num1,num2):
        print (f'\nDifference: {num1-num2}')
    def multiply (num1,num2):
        print (f'\nProduct: {num1*num2}')
    def divide (num1,num2):
        print (f'\nQuotient: {num1/num2}')
    def square (num1):
        print(f'\nSquare: {num1**2}')
    def power (num1,num2):
        print(f'\nPower: {num1**num2}')
    def squareroot (num1):
        print(f'\nSquare Root: {sqrt(num1)}')
    def log(num1):
        print(f'\nLog base 10: {log10(num1)}')
    def sin(num1):
        print(f'\nSine in radians: {sin(num1)}')
    def cos(num1):
        print(f'\nCosine in radians: {cos(num1)}')
    def tan(num1):
        print(f'\nTangent in radians: {tan(num1)}')
    def factorial(num1):
        print(f'\nFactorial: {factorial(num1)}')
    def frtode(num1,num2):
        print(f'\nFraction to Decimal: {num1/num2}')
    def gcd(num1,num2):
        print(f'\nGreatest Common Denominator: {gcd(num1,num2)}')
    def average(num1,num2):
        print(f'\nAverage: {(num1+num2)/2}')
    
print('\nChoose from the following options: ')
print('1. Addition')
print('2. Subtraction')
print('3. Multiplication')
print('4. Division')
print('5. Square')
print('6. Power')
print('7. Square Root')
print('8. Log base 10')
print('9. Sine in radians')
print('10. Cosine in radians')
print('11. Tangent in radians')
print('12. Factorial')
print('13. Fraction to decimal')
print('14. Greatest common denominator')
print('15. Average')
print('16. Exit')

while True:
    choice=int(input('\nEnter your choice (1-16): '))
    
    if choice==1:
        num1=int(input('Enter the first number: '))
        num2=int(input('Enter the second number: '))
        calc.addition (num1,num2)
    elif choice==2:
        num1=int(input('Enter the first number: '))
        num2=int(input('Enter the second number: '))
        calc.subtract (num1,num2)
    elif choice==3:
        num1=int(input('Enter the first number: '))
        num2=int(input('Enter the second number: '))
        calc.multiply (num1,num2)
    elif choice==4:
        num1=int(input('Enter the first number: '))
        num2=int(input('Enter the second number: '))
        if num2==0:
            print('Error')
        else:
            calc.divide(num1,num2)
    elif choice==5:
        num1=int(input('Enter the number: '))
        calc.square(num1)
    elif choice==6:
        num1=int(input('Enter the base number: '))
        num2=int(input('Enter the power: '))
        calc.power (num1,num2)
    elif choice==7:
        num1=int(input('Enter the number: '))
        if num1<0:
            print('Error')
        else:
            calc.squareroot (num1)
    elif choice==8:
        num1=int(input('Enter the number: '))
        calc.log (num1)
    elif choice==9:
        num1=int(input('Enter the number: '))
        calc.sin (num1)
    elif choice==10:
        num1=int(input('Enter the number: '))
        calc.cos (num1)
    elif choice==11:
        num1=int(input('Enter the number: '))
        calc.tan (num1)
    elif choice==12:
        num1=int(input('Enter the number: '))
        calc.factorial (num1)
    elif choice==13:
        num1=int(input('Enter the numerator: '))
        num2=int(input('Enter the denominator: '))
        calc.frtode (num1,num2)
    elif choice==14:
        num1=int(input('Enter the first number: '))
        num2=int(input('Enter the second number: '))
        calc.gcd (num1,num2)
    elif choice==15:
        num1=int(input('Enter the first number: '))
        num2=int(input('Enter the second number: '))
        calc.average (num1,num2)
    elif choice==16:
        break
    else:
        print('Invalid input')
```
